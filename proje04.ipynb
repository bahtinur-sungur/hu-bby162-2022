```python
from google.colab import drive

kayitDosya = "/content/drive/MyDrive/Colab Notebooks/veritabani.txt "

def kayitlariListele():
  katalog = open(kayitDosya, "r")
  for line in katalog.readlines():
    print(line)
  katalog.close()
  anaMenu()

def yeniKayitEkle():
  kitapAdi = input("Kitap adı giriniz: ")
  yazarAdi = input("Yazar adı giriniz: ")
  yayinevi = input("Yayınevi giriniz: ")
  basimYili = input("Basım yılı giriniz: ")
  f = open(kayitDosya, 'a')
  f.write(kitapAdi + "|")
  f.write(yazarAdi + "|")
  f.write(yayinevi + "|")
  f.write(basimYili + "\n")
  f.close()
  print("Tebrikler :) Kaydınız başarıyla tamamlandı.")
  anaMenu()
 
def kayitAra():
  print("Lütfen sadece kitap adı giriniz")
  ara = input("Arama yapmak istediğiniz kayıt adı: ")
  kayit = open(kayitDosya, "r")

  aranan = kayit.read().find(ara)
  if aranan:
    print("Aranan kayıt kütüphane kataloğunda mevcuttur.")
    anaMenu()
  else:
    print("Aranan kayıt kütüphane kataloğunda mevcut değildir.")
    anaMenu()

def cikis():
  print("\t\t----------\nKütüphane kataloğundan çıkış yapmak istiyor musunuz?\n  \t1) Evet\n  \t2) Hayır\n\t\t----------")
  cikisIslem = input("İşlem seçiniz: ")
  if cikisIslem == "2":
    print("Ana menüye yönlendiriliyorsunuz. Lütfen bir seçim yapınız.")
    anaMenu()
  elif cikisIslem == "1":
   print("Çıkış yaptınız. İyi günler")
  else:
    print("Hatalı işlem gerçekleştirdiniz. Seçimlerinizi sadece 1 veya 2 olarak giriniz.")
    cikis()

def kayitlariSil():
  open(kayitDosya, 'w').close()
  print("Kayıtlar silindi, ana menüye yönlendiriliyorsunuz. İşleminize yeni kayıt ekleme ile başlayabilirsiniz, kolay gelsin :)")
  anaMenu()

def anaMenu():
  print("\t\t----------\nLütfen yapacağınız işlemi seçiniz\n  \t1) Kayıt Listele\n  \t2) Yeni Kayıt Ekle\n  \t3) Kayıt Ara\n  \t4) Kayıtları Sil (Seçim Yaparken Lütfen Dikkat Ediniz. Tüm Kayıtlar Silinmektedir.)\n  \t5) Çıkış Yap\n\t\t----------")
  islem = input("İşlem seçiniz: ")
  if islem == "1":
    kayitlariListele()
  elif islem == "2":
    yeniKayitEkle()
  elif islem== "3":
    kayitAra()
  elif islem == "4":
    kayitlariSil()
  elif islem == "5":
    cikis()
  else:
    print("Hatalı giriş yaptınız, ana menüye tekrar yönlendiriliyorsunuz...")
    anaMenu()

print("KÜTÜPHANE KATALOĞUNA HOŞGELDİNİZ")
anaMenu()
```
